```python
import pandas as pd
names = ['Bob','Jessica','Mary','John','Mel']
grades = [76,83,77,78,95]
GradeList = zip(names,grades)
df = pd.DataFrame(data = GradeList,
columns=['Names', 'Grades'])
%matplotlib inline
df.plot()
```

```python
import matplotlib.pyplot as plt
df.plot()
displayText = "Wow"
xloc = 1
yloc = 76
xtext = 8
ytext = 0
plt.annotate(displayText,xy=(xloc, yloc),xytext=(xtext,ytext),xycoords=('axes fraction', 'data'),textcoords='offset points')
```

```python
import matplotlib.pyplot as plt
df.plot()
displayText = "WOW"
xloc = 1
yloc = 76
xtext = 8
ytext = -50
plt.annotate(displayText, xy=(xloc, yloc),arrowprops=dict(facecolor='black',shrink=0.08),
             xytext=(xtext,ytext),xycoords=('axes fraction', 'data'),textcoords='offset points')
```
